```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
```

```python



# Creating a numpy array
altezza = np.array([160,165,170,175,175,178,190,192,150]).reshape((-1, 1))
peso = np.array([62,60,68,70,80,82,92,98,61])

# Plotting point using scatter method
plt.scatter(altezza,peso)
model = LinearRegression()
model = model.fit(altezza, peso)
peso_pred = model.predict(altezza)
plt.xlabel("Altezza")
plt.ylabel("Peso")
r_sq = model.score(altezza, peso)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)


h_new = np.linspace(140, 200)
p_new = model.predict(h_new[:, np.newaxis])
ax = plt.axes()
ax.plot(h_new, p_new)
plt.show()



# Creating a numpy array
altezza = np.array([160,165,170,175,175,178,190,192,150]).reshape((-1, 1))
peso = np.array([62,60,68,70,80,82,92,98,61])

# Plotting point using scatter method
plt.scatter(altezza,peso)
model = LinearRegression()
model = model.fit(altezza, peso)
peso_pred = model.predict(altezza)
plt.xlabel("Altezza")
plt.ylabel("Peso")
r_sq = model.score(altezza, peso)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)


h_new = np.linspace(140, 200)
p_new = model.predict(h_new[:, np.newaxis])
ax = plt.axes()
ax.plot(h_new, p_new)
plt.show()
# plot the results




df = pd.read_csv ('dataset.csv')
superficie = np.array(df['m2superficie']).reshape((-1, 1))
prezzo=np.array((df['prezzo']))
plt.scatter(superficie,prezzo)
plt.xlabel("Superficie m2")
plt.ylabel("Prezzo")
model = LinearRegression()
model = model.fit(superficie, prezzo)
prezzo_pred = model.predict(superficie)
super_new = np.linspace(30, 200)
prezzo_new = model.predict(super_new[:, np.newaxis])
ax = plt.axes()
ax.get_yaxis().get_major_formatter().set_scientific(False)

ax.plot(super_new, prezzo_new)
plt.show()




```
